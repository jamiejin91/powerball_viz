```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import requests
import json
```

```python
# url that holds dates in range wanted
url = "http://www.powerball.com/powerball/pb_nbr_history.asp?startDate=02%2F03%2F2010&endDate=11%2F29%2F2017"
```

```python
#get response from site
response = requests.get(url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
results = soup.find_all('td')[43:9844]
```

```python
results[-10:-1]
```

```python
lotto_numbers = []
count = 0

for drawing in results:
    if drawing.b:
        date_drawing = {}
        date_drawing['date'] = drawing.b.text
        numbers = []
        powerball = 0
    elif (len(numbers) < 6):
        numbers.append(drawing.text)
    elif (len(numbers) == 6 and powerball == 0):
        date_drawing['numbers'] = numbers[0:5]
        powerball = drawing.text
#         print(drawing.text)
        date_drawing['powerball'] = powerball
        lotto_numbers.append(date_drawing)
        

lotto_numbers
        
        
```

```python
drawing_df = pd.DataFrame(lotto_numbers)
drawing_df.tail()
```
